```python
# 2번, 5번 대분류를 예측하는것이 강건
# 2번 5번 대분류에 속한 중분류 중, 판매량이 상위 1개인 중분류의 소분류
# 52, 53, 28, 41
```

```python
import pandas as pd

# 파일 경로 패턴에 맞춰 모든 파일을 읽어옵니다.
frames = [pd.read_csv(f'/content/data/sub/B002-C003-00{i:02}_data.csv') for i in range(1, 54)]

# 모든 데이터프레임을 하나로 합칩니다.
result = pd.concat(frames)

# 1 labeling
result['baby'] = result['소분류'].apply(lambda x: 1 if x in ['B002-C003-0052', 'B002-C003-0053', 'B002-C003-0028', 'B002-C003-0041'] else 0)

# 결과를 새 CSV 파일에 저장합니다.
result.to_csv('/content/data/baby.csv', index=False)
```
